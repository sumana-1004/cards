# Electricity Bill Generator (Slab-Based, Multiple Households)

The Electricity Board wants to calculate electricity bills for multiple households.

Store the electricity tariff slabs in a tuple.

Units	Rate (₹/unit),

First 100	₹1.50,

Next 100	₹2.50,

Next 300	₹4.00,

Above 500	₹6.00,



The program should:

i) Ask the user for the number of households.

ii) Store the household names in a list.

iii) Store the units consumed by each household in another list.

iv) Calculate the electricity bill according to the slab rates.
Display the details in a neat table.


If the PAN number starts with 'A', add an additional charge of ₹8 × Units Consumed.

```python
households_no = int(input("Number of Households: "))
households_list = []
units_list = []
total_bill_list = []
PAN_list = []
Name_list = []
bill_values_list = []
extra_charge_values_list = []

for i in range(households_no):
  print(f"\nEnter the details for Households {i+1}")
  Name = input("Household Name: ")
  PAN = input("PAN Number: ")
  Units = int(input("Units Consumed: "))

  units_list.append(Units)
  households_list.append(Name)
  PAN_list.append(PAN)
  Name_list.append(Name)

  bill = 0
  if Units <= 100:
    bill = Units * 1.5
  elif Units <= 200:
    bill = (Units - 100)*2.5 + 100*1.5
  elif Units <= 500:
    bill = (Units - 200)*4.0 + 100*2.5 + (100*1.5)
  elif Units > 500:
    bill = (Units - 500)*6.0 + (300*4.0) + (100*2.5) + (100*1.5)

  extra_charge = 0
  if PAN[0] == 'A':
    extra_charge = 8*Units

  bill_values_list.append(bill)
  extra_charge_values_list.append(extra_charge)

  total_bill = bill + extra_charge
  total_bill_list.append(total_bill)

print("-----------------------------------------------------")
print("Household\tPAN Number\tUnits\tBill (Rs.)\tExtra Charge (Rs.)\tTotal Bill (Rs.)")
for i in range(households_no):
  print(f"{Name_list[i]}\t\t{PAN_list[i]}\t{units_list[i]}\t{bill_values_list[i]}\t\t{extra_charge_values_list[i]}\t\t\t{total_bill_list[i]}")
print("-----------------------------------------------------")

print(f"Highest Bill: {Name_list[total_bill_list.index(max(total_bill_list))]} ({max(total_bill_list):.2f})")
print(f"Lowest Bill: {Name_list[total_bill_list.index(min(total_bill_list))]} ({min(total_bill_list):.2f})")
print(f"Total Revenue: ", sum(total_bill_list))
print(f"Average Bill: ", sum(total_bill_list)/households_no)
```

# Question 1: Smart Supermarket Billing & Customer Analytics
## Problem Statement:
A supermarket records purchases made by multiple customers. For each customer, collect:  

*   Customer ID
*   Customer Name
*   PAN Number
*   Purchased Items (multiple items)
*   Quantity of each item
*   Unit Price of each item

## Requirements
- Calculate the total bill for each customer.  
- Apply the following discounts:  
  * 5% if the bill exceeds ₹5,000.  
  * Additional 3% if the customer purchases more than 10 items.  
  * If the PAN number starts with A, provide a cashback of ₹8 for every item purchased.  
  * Store customer information using appropriate Python data structures.  

## Identify:  
* Customer with the highest bill.  
* Most frequently purchased item.  
* Item purchased by the maximum number of unique customers.  
* Customers who purchased exactly the same set of items.  

## Generate a final report showing:  
  - Total sales  
  - Average customer bill  
  - Total discount provided  
  - Top 3 customers based on bill amount  

```python
customers = []
another_customer = "Y"

while another_customer.lower() in ("y", "yes"):

    customer_id = input("Enter Customer ID: ")
    customer_name = input("Enter your name: ")
    pan_number = input("Enter your PAN number: ")

    items = []
    another_item = "Y"

    while another_item.lower() in ("y", "yes"):
        item_name = input("Enter the item you wish to buy: ")
        item_quantity = int(input("Enter the quantity of the item: "))
        item_price = float(input("Enter the price of the item: "))

        item = {
            "name": item_name,
            "quantity": item_quantity,
            "price": item_price
        }

        items.append(item)
        another_item = input("Add another item for this customer? (Y/N): ")

    gross_bill = 0.0
    total_quantity = 0

    for item in items:
        gross_bill += item["quantity"] * item["price"]
        total_quantity += item["quantity"]

    discount_amount = 0.0

    if gross_bill > 5000:
        discount_amount += 0.05 * gross_bill

    if total_quantity > 10:
        discount_amount += 0.03 * gross_bill

    bill_after_discount = gross_bill - discount_amount

    cashback = 0.0
    if pan_number.startswith("A"):
        cashback = 8 * total_quantity

    net_bill = bill_after_discount - cashback

    customer = {
        "id": customer_id,
        "name": customer_name,
        "pan": pan_number,
        "items": items,
        "gross_bill": gross_bill,
        "total_quantity": total_quantity,
        "discount": discount_amount,
        "cashback": cashback,
        "net_bill": net_bill
    }

    customers.append(customer)

    print("-----------Customer Bill--------------")
    print("Customer ID:", customer_id)
    print("Customer Name:", customer_name)
    print("Gross Bill:", gross_bill)
    print("Total Quantity:", total_quantity)
    print("Discount:", discount_amount)
    print("Cashback:", cashback)
    print("Net Bill:", net_bill)
    print("--------------------------------------")

    another_customer = input("Add another customer? (Y/N): ")

print("==========CUSTOMERS SUMMARY===========")
for c in customers:
    print(f"Customer ID: {c['id']}")
    print(f"Customer Name: {c['name']}")
    print(f"Gross Bill: {c['gross_bill']}")
    print(f"Discount: {c['discount']}")
    print(f"Cashback: {c['cashback']}")
    print(f"Net Bill: {c['net_bill']}")
    print("--------------------------------------")

if customers:
    highest_customer = customers[0]
    for c in customers:
        if c["net_bill"] > highest_customer["net_bill"]:
            highest_customer = c

    print("===== Customer with Highest Bill =====")
    print("Customer ID:", highest_customer["id"])
    print("Customer Name:", highest_customer["name"])
    print("Net Bill:", highest_customer["net_bill"])
    print("--------------------------------------")
else:
    print("No customers to analyze.")

item_quantity_map = {}

for c in customers:
    for item in c["items"]:
        name = item["name"]
        qty = item["quantity"]

        if name in item_quantity_map:
            item_quantity_map[name] += qty
        else:
            item_quantity_map[name] = qty

if item_quantity_map:
    most_freq_item = None
    most_freq_qty = 0

    for name, qty in item_quantity_map.items():
        if qty > most_freq_qty:
            most_freq_qty = qty
            most_freq_item = name

    print("===== Most Frequently Purchased Item =====")
    print("Item Name:", most_freq_item)
    print("Total Quantity Purchased:", most_freq_qty)
    print("------------------------------------------")
else:
    print("No items to analyze.")

item_to_customers = {}

for c in customers:
    cid = c["id"]
    customer_item_names = set(item["name"] for item in c["items"])

    for name in customer_item_names:
        if name not in item_to_customers:
            item_to_customers[name] = set()
        item_to_customers[name].add(cid)

if item_to_customers:
    max_item = None
    max_unique_customers = 0

    for name, cust_set in item_to_customers.items():
        count = len(cust_set)
        if count > max_unique_customers:
            max_unique_customers = count
            max_item = name

    print("===== Item Bought by Max Unique Customers =====")
    print("Item Name:", max_item)
    print("Number of Unique Customers:", max_unique_customers)
    print("-----------------------------------------------")
else:
    print("No items to analyze for unique customers.")

from collections import defaultdict

pattern_to_customers = defaultdict(list)

for c in customers:
    item_names_set = set(item["name"] for item in c["items"])
    item_pattern = tuple(sorted(item_names_set))
    pattern_to_customers[item_pattern].append(c)

print("===== Customers with Same Set of Items =====")
found_group = False
for pattern, cust_list in pattern_to_customers.items():
    if len(cust_list) > 1:
        found_group = True
        print("Item Set:", pattern)
        print("Customers:")
        for cust in cust_list:
            print(f"  ID: {cust['id']}, Name: {cust['name']}")
        print("--------------------------------------")

if not found_group:
    print("No customers share exactly the same set of items.")

total_sales = sum(c["net_bill"] for c in customers)
total_discount = sum(c["discount"] for c in customers)

if customers:
    average_bill = total_sales / len(customers)
else:
    average_bill = 0.0

top_customers = sorted(customers, key=lambda c: c["net_bill"], reverse=True)[:3]

print("========== FINAL REPORT ==========")
print("Total Sales:", total_sales)
print("Average Customer Bill:", average_bill)
print("Total Discount Provided:", total_discount)
print("Top 3 Customers (by Net Bill):")
for c in top_customers:
    print(f"  ID: {c['id']}, Name: {c['name']}, Net Bill: {c['net_bill']}")
print("==================================")
```

# Question 2: University Examination Analytics
## Problem Statement:
A university wants to analyze examination results of students from different departments.
For each student, collect:
* Register Number  
* Name  
* Department  
* Marks in five subjects  

## Requirements
Compute total, average, percentage, grade, and pass/fail status.  

Store:  
* Student identity as a tuple.  
* Student records in a dictionary.  
* Department-wise students using sets.  

Determine:  
* Overall topper.  
* Department-wise toppers.  
* Students scoring above 90 in at least three subjects.  
* Students failing in exactly one subject.  
* Subject with the highest class average.  
* Detect duplicate student names across departments.  

Generate a department-wise report containing:  
* Number of students  
* Pass percentage  
* Average marks  
* Highest and lowest scorer  

```python
students = []
departments = set()
dept_students = {}

another_student = "Y"

while another_student.lower() in ("y", "yes"):
    reg_no = input("Enter Register Number: ")
    name = input("Enter Name: ")
    dept = input("Enter Department: ")

    marks = []
    for i in range(5):
        m = int(input(f"Enter marks for subject {i + 1}: "))
        marks.append(m)

    total = sum(marks)
    average = total / 5
    percentage = (total / 500) * 100

    if percentage >= 90:
        grade = "S"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    elif percentage >= 50:
        grade = "D"
    else:
        grade = "F"

    pass_fail = "Pass"
    for m in marks:
        if m < 50:
            pass_fail = "Fail"
            break

    identity = (reg_no, name, dept)

    record = {
        "identity": identity,
        "marks": marks,
        "total": total,
        "average": average,
        "percentage": percentage,
        "grade": grade,
        "status": pass_fail
    }

    students.append(record)

    departments.add(dept)
    if dept not in dept_students:
        dept_students[dept] = set()
    dept_students[dept].add(identity)

    print("-----------Student Result--------------")
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Marks:", marks)
    print("Total:", total)
    print("Average:", average)
    print("Percentage:", percentage)
    print("Grade:", grade)
    print("Status:", pass_fail)
    print("--------------------------------------")

    another_student = input("Add another student? (Y/N): ")

print("==========STUDENTS SUMMARY===========")
for s in students:
    reg_no, name, dept = s["identity"]
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Marks:", s["marks"])
    print("Total:", s["total"])
    print("Average:", s["average"])
    print("Percentage:", s["percentage"])
    print("Grade:", s["grade"])
    print("Status:", s["status"])
    print("--------------------------------------")

if students:
    overall_topper = students[0]
    for s in students:
        if s["total"] > overall_topper["total"]:
            overall_topper = s

    reg_no, name, dept = overall_topper["identity"]
    print("===== Overall Topper =====")
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Total Marks:", overall_topper["total"])
    print("--------------------------------------")

dept_toppers = {}

for s in students:
    reg_no, name, dept = s["identity"]
    if dept not in dept_toppers:
        dept_toppers[dept] = s
    else:
        if s["total"] > dept_toppers[dept]["total"]:
            dept_toppers[dept] = s

print("===== Department-wise Toppers =====")
for dept, topper in dept_toppers.items():
    reg_no, name, d = topper["identity"]
    print("Department:", dept)
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Total Marks:", topper["total"])
    print("--------------------------------------")

above_90_three = []

for s in students:
    count_90 = 0
    for m in s["marks"]:
        if m >= 90:
            count_90 += 1
    if count_90 >= 3:
        above_90_three.append(s)

print("===== Students with >=90 in at least 3 subjects =====")
if above_90_three:
    for s in above_90_three:
        reg_no, name, dept = s["identity"]
        print("Register Number:", reg_no)
        print("Name:", name)
        print("Department:", dept)
        print("Marks:", s["marks"])
        print("--------------------------------------")
else:
    print("None")
    print("--------------------------------------")

fail_one = []

for s in students:
    fail_count = 0
    for m in s["marks"]:
        if m < 50:
            fail_count += 1
    if fail_count == 1:
        fail_one.append(s)

print("===== Students failing in exactly one subject =====")
if fail_one:
    for s in fail_one:
        reg_no, name, dept = s["identity"]
        print("Register Number:", reg_no)
        print("Name:", name)
        print("Department:", dept)
        print("Marks:", s["marks"])
        print("--------------------------------------")
else:
    print("None")
    print("--------------------------------------")

subject_totals = [0, 0, 0, 0, 0]
subject_counts = [0, 0, 0, 0, 0]

for s in students:
    for i in range(5):
        subject_totals[i] += s["marks"][i]
        subject_counts[i] += 1

subject_averages = []
for i in range(5):
    if subject_counts[i] > 0:
        avg = subject_totals[i] / subject_counts[i]
    else:
        avg = 0
    subject_averages.append(avg)

highest_avg_subject = 0
for i in range(5):
    if subject_averages[i] > subject_averages[highest_avg_subject]:
        highest_avg_subject = i

print("===== Subject with Highest Class Average =====")
print("Subject Number:", highest_avg_subject + 1)
print("Average Marks:", subject_averages[highest_avg_subject])
print("--------------------------------------")

name_depts = {}

for s in students:
    reg_no, name, dept = s["identity"]
    if name not in name_depts:
        name_depts[name] = set()
    name_depts[name].add(dept)

duplicate_names = []

for name, depts in name_depts.items():
    if len(depts) > 1:
        duplicate_names.append((name, depts))

print("===== Duplicate Student Names Across Departments =====")
if duplicate_names:
    for name, depts in duplicate_names:
        print("Name:", name)
        print("Departments:", list(depts))
        print("--------------------------------------")
else:
    print("None")
    print("--------------------------------------")

print("===== Department-wise Report =====")
for dept in departments:
    dept_list = []
    for s in students:
        reg_no, name, d = s["identity"]
        if d == dept:
            dept_list.append(s)

    if not dept_list:
        continue

    num_students = len(dept_list)

    pass_count = 0
    total_marks_sum = 0
    highest_student = dept_list[0]
    lowest_student = dept_list[0]

    for s in dept_list:
        total_marks_sum += s["total"]
        if s["status"] == "Pass":
            pass_count += 1
        if s["total"] > highest_student["total"]:
            highest_student = s
        if s["total"] < lowest_student["total"]:
            lowest_student = s

    pass_percentage = (pass_count / num_students) * 100
    average_marks = total_marks_sum / num_students

    reg_h, name_h, d_h = highest_student["identity"]
    reg_l, name_l, d_l = lowest_student["identity"]

    print("Department:", dept)
    print("Number of Students:", num_students)
    print("Pass Percentage:", pass_percentage)
    print("Average Marks:", average_marks)
    print("Highest Scorer:", name_h, "(Reg:", reg_h, ", Total:", highest_student["total"], ")")
    print("Lowest Scorer:", name_l, "(Reg:", reg_l, ", Total:", lowest_student["total"], ")")
    print("--------------------------------------")
```

## Question 4: College Course Registration System

Problem Statement
- Students register for elective courses.

For each student collect:

* Register Number  
* Name  
* Department  
* List of electives selected  
* Requirements
* A student cannot register for the same course twice.  

Find:  
* Most popular elective.  
* Least popular elective.  
* Students choosing more than three electives.  
* Students with identical elective combinations.  
* Electives selected by only one student.  
* Display department-wise enrollment statistics.  
* Recommend one additional elective for every student based on the electives chosen by other students in the same department.  

Display all reports in sorted order.  

```python
students = []
departments = set()

another_student = "Y"

while another_student.lower() in ("y", "yes"):
    reg_no = input("Enter Register Number: ")
    name = input("Enter Name: ")
    dept = input("Enter Department: ")

    electives = []
    another_elective = "Y"

    while another_elective.lower() in ("y", "yes"):
        elective = input("Enter elective name: ")

        if elective in electives:
            print("You have already selected this elective. Choose a different one.")
        else:
            electives.append(elective)

        another_elective = input("Add another elective? (Y/N): ")

    identity = (reg_no, name, dept)

    record = {
        "identity": identity,
        "electives": electives
    }

    students.append(record)
    departments.add(dept)

    print("-----------Student Registration--------------")
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Electives:", electives)
    print("---------------------------------------------")

    another_student = input("Add another student? (Y/N): ")

print("==========STUDENTS SUMMARY===========")
for s in students:
    reg_no, name, dept = s["identity"]
    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Electives:", s["electives"])
    print("--------------------------------------")

elective_counts = {}

for s in students:
    for e in s["electives"]:
        if e in elective_counts:
            elective_counts[e] += 1
        else:
            elective_counts[e] = 1

if elective_counts:
    popular_electives = sorted(elective_counts.items(), key=lambda x: x[1], reverse=True)
    most_popular = popular_electives[0]
    least_popular = popular_electives[-1]

    print("===== Most Popular Elective =====")
    print("Elective:", most_popular[0])
    print("Number of Students:", most_popular[1])
    print("---------------------------------")

    print("===== Least Popular Elective =====")
    print("Elective:", least_popular[0])
    print("Number of Students:", least_popular[1])
    print("---------------------------------")
else:
    print("No electives to analyze.")

students_more_than_three = []

for s in students:
    if len(s["electives"]) > 3:
        students_more_than_three.append(s)

students_more_than_three = sorted(students_more_than_three, key=lambda x: x["identity"][0])

print("===== Students Choosing More Than Three Electives =====")
if students_more_than_three:
    for s in students_more_than_three:
        reg_no, name, dept = s["identity"]
        print("Register Number:", reg_no)
        print("Name:", name)
        print("Department:", dept)
        print("Electives:", s["electives"])
        print("---------------------------------")
else:
    print("None")
    print("---------------------------------")

from collections import defaultdict

pattern_to_students = defaultdict(list)

for s in students:
    reg_no, name, dept = s["identity"]
    elective_set = tuple(sorted(s["electives"]))
    pattern_to_students[elective_set].append(s)

print("===== Students with Identical Elective Combinations =====")
found_pattern = False
for pattern in sorted(pattern_to_students.keys()):
    student_list = pattern_to_students[pattern]
    if len(student_list) > 1:
        found_pattern = True
        print("Elective Combination:", list(pattern))
        for s in sorted(student_list, key=lambda x: x["identity"][0]):
            reg_no, name, dept = s["identity"]
            print("  Register Number:", reg_no, "Name:", name, "Department:", dept)
        print("---------------------------------")

if not found_pattern:
    print("None")
    print("---------------------------------")

electives_only_one = []

for e, count in elective_counts.items():
    if count == 1:
        electives_only_one.append(e)

electives_only_one = sorted(electives_only_one)

print("===== Electives Selected by Only One Student =====")
if electives_only_one:
    for e in electives_only_one:
        print("Elective:", e)
    print("---------------------------------")
else:
    print("None")
    print("---------------------------------")

dept_enrollment = {}

for dept in departments:
    dept_enrollment[dept] = []

for s in students:
    reg_no, name, dept = s["identity"]
    dept_enrollment[dept].append(s)

print("===== Department-wise Enrollment Statistics =====")
for dept in sorted(departments):
    dept_list = dept_enrollment[dept]
    total_students = len(dept_list)
    total_electives = 0
    for s in dept_list:
        total_electives += len(s["electives"])

    if total_students > 0:
        avg_electives = total_electives / total_students
    else:
        avg_electives = 0

    print("Department:", dept)
    print("Number of Students:", total_students)
    print("Average Number of Electives:", avg_electives)
    print("---------------------------------")

recommendations = {}

for dept in departments:
    dept_list = dept_enrollment[dept]
    dept_elective_counts = {}

    for s in dept_list:
        for e in s["electives"]:
            if e in dept_elective_counts:
                dept_elective_counts[e] += 1
            else:
                dept_elective_counts[e] = 1

    sorted_dept_electives = sorted(dept_elective_counts.items(), key=lambda x: x[1], reverse=True)
    recommendations[dept] = [e for e, c in sorted_dept_electives]

print("===== Recommended Additional Elective for Each Student =====")
students_sorted = sorted(students, key=lambda x: x["identity"][0])

for s in students_sorted:
    reg_no, name, dept = s["identity"]
    current_electives = set(s["electives"])
    recommended = None

    for e in recommendations[dept]:
        if e not in current_electives:
            recommended = e
            break

    print("Register Number:", reg_no)
    print("Name:", name)
    print("Department:", dept)
    print("Current Electives:", s["electives"])
    if recommended is not None:
        print("Recommended Elective:", recommended)
    else:
        print("Recommended Elective: None (already has all common electives)")
    print("---------------------------------")
```
